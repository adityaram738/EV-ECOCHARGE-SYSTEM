```python
import pandas as pd

fleet_df = pd.read_csv('fleet_data.csv')
prices_df = pd.read_csv('electricity_prices.csv')

# Sort prices (cheapest first)
prices_df = prices_df.sort_values(by="Price_per_kWh")

def calculate_schedule(fleet_df, prices_df):
    schedule = []

    for _, row in fleet_df.iterrows():
        vehicle = row['Vehicle_ID']
        needed = row['Required_Battery_Pct'] - row['Current_Battery_Pct']
        
        # Assume: 20% battery = 1 hour charging
        hours_needed = max(1, needed // 20)

        cheapest_hours = prices_df.head(hours_needed)['Hour'].tolist()
        
        schedule.append({
            "Vehicle": vehicle,
            "Charge Hours": hours_needed,
            "Best Time": cheapest_hours
        })

    return pd.DataFrame(schedule)

result = calculate_schedule(fleet_df, prices_df)
result
```
